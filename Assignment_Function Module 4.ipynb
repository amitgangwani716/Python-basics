Q3. What are the different ways to define and call a function in Python?

Ans3. 1. Regular Function Definition
Definition:


def greet(name):
    return f"Hello, {name}!"
Calling:


message = greet("Alice")

2. Lambda Function
Definition:


greet = lambda name: f"Hello, {name}!"
Calling:


message = greet("Bob")
print(message)  # Output: Hello, Bob!
3. Function with Default Arguments
Definition:


def greet(name="Guest"):
    return f"Hello, {name}!"
Calling:


message1 = greet("Charlie")
message2 = greet()
print(message1)  # Output: Hello, Charlie!
print(message2)  # Output: Hello, Guest!
4. Function with Variable-Length Arguments
Definition:


def greet(*names):
    return "Hello, " + ", ".join(names) + "!"
Calling:


message = greet("Dave", "Eve", "Frank")
print(message)  # Output: Hello, Dave, Eve, Frank!
5. Function with Keyword Arguments
Definition:


def greet(**kwargs):
    return f"Hello, {kwargs.get('name', 'Guest')}!"
Calling:


message = greet(name="Grace")
print(message)  # Output: Hello, Grace!
6. Nested Functions
Definition:


def outer_function(name):
    def inner_function():
        return f"Hello, {name}!"
    return inner_function()
Calling:


message = outer_function("Hank")
print(message)  # Output: Hello, Hank!
7. Higher-Order Functions
Definition:


def greet(name):
    return f"Hello, {name}!"

def call_function(func, name):
    return func(name)
Calling:


message = call_function(greet, "Ivy")
print(message)  # Output: Hello, Ivy!

These are some of the versatile ways to define and call functions in Python. Each method has its own use case and can be chosen based on the specific requirements of your code.

```python
#Ans1 
def sum_of_even_number(number):
    return sum(num for num in number if num % 2==0)

number = [1,2,3,4,5,6]
print (sum_of_even_number(number))
```

3. Implement a Python function that takes a list of integers and returns a new list containing the squares of
each number

```python
#Ans3. 

def square_numbers(nums):
    squared_nums = []
    for num in nums:
        squared_nums.append(num**2)
    return squared_nums

# Example usage:
numbers = [1, 2, 3, 4, 5]
squared_numbers = square_numbers(numbers)
print(squared_numbers)
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
#Ans4. 

def is_prime(n):
    """Check if a number is prime."""
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# Example usage:
for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is a prime number.")
    else:
        print(f"{num} is not a prime number.")
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent

```python
#Ans6. 

def powers_of_two(max_exponent):
    exponent = 0
    while exponent <= max_exponent:
        yield 2 ** exponent
        exponent += 1

# Example usage:
for power in powers_of_two(6):
    print(power)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
#Ans9. # Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 20, 37, 100]

# Using map to convert the list of Celsius temperatures to Fahrenheit
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Print the converted temperatures
print("Temperatures in Fahrenheit:", fahrenheit_temperatures)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
#Ans10.

def remove_vowels(input_string):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda char: char not in vowels, input_string))

# Example usage
input_string = "Hello, World!"
result = remove_vowels(input_string)
print("Original string:", input_string)
print("String without vowels:", result)
```
